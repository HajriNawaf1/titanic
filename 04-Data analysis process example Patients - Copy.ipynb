```python
# Your cleaning code here
#'California','New York','Illinois': 'Florida': 'FL','Nebraska': 'NE'

df_clean['state'].replace( { 'California' :'CA' , 'New York': 'NY', 
                            'Illinois': 'IL' ,'Florida' :'FL'
                            , 'Nebraska':'NE'} ,inplace=True)
```

# Q1 : does higher weight means you will be patient ?

```python
#Avg
mean = df_clean['weight'].mean()
mean
```

```python
#high

high_weigh = df_clean.loc[ df_clean['weight'] >= mean ]
high_weigh
```

```python
#low

low_weigh = df_clean.loc[ df_clean['weight'] < mean ]
low_weigh
```

```python
x = high_weigh.shape[0]
```

```python
# get percentage 

high_weigh_perc = (x /df_clean.shape[0])*100

low_weigh_perc = (low_weigh.shape[0]/df_clean.shape[0])*100
```

```python
low_weigh_perc
```

```python
high_weigh_perc
```
