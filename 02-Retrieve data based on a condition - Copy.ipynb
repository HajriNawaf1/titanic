## 2: HOW TO USE LOC (location)

```python
# How to retrive based on one condition

df.loc[ df['Age'] > 10 ]
```

```python
## How to retrive based on Multiple conditions.

df.loc[ (df['Age'] > 10 ) & (df['Age'] <50) ]
```

```python
df.loc[ ((df['Age'] > 10 ) | (df['Age'] <50)) & (df['Survived'] == 1)]
```

```python
# People who survied : 

survied_df = df.loc[df['Survived'] == 1]
```

```python
survied_df['Age'].median()
```

## 3: HOW TO USE query

```python
# How to retrive based on one conditions using Query.

df.query('Pclass == 1')
```

```python
# How to retrive based on multiple conditions using Query.
df.query('Age > 10 and Pclass == 1')
```

## 4: HOW TO USE nlargest and nsmallest

```python
# How to retrive the 10 oldest people in the ship

df.nlargest(10,'Age')
```

```python
# How to retrive the 10 youngest people in the ship

df.nsmallest(10,'Age')
```

```python
# How to sort values in the dataset based on speceific column.

df.sort_values('Name', ascending= True)
```

# Conditional change

### option 1: 

### what is the number of people in each class ?

```python
# Retrive the number of people in each class ?

df['Pclass'].value_counts()
```

```python
# Change values 

#df.loc[df['Pclass'] == 1]

df.loc[df['Pclass'] == 1 , 'Pclass'] = 'VIP'
```

```python
df.loc[df['Pclass'] == 'VIP' , 'Survived'] = 1
```

```python
# Show the resault 

df.head(10)
```

```python
# Can i change more than one column ? 

df.loc[df['Pclass'] == 'VIP' , ['Pclass' , 'Survived']] = [1 , 0]
```

```python
# Show the resault 

df.head(10)
```

```python
# Head()
```

```python
# Can I change for range of values ? EX: age
```

```python
df.loc[(df['Age'] >= 60 ) & (df['Age'] <= 80) , 'Survievd'] = 0
```

```python
df.head()
```

```python
df.loc[df['Pclass'] == 1 , 'Pclass' ] = 'VIP'
df.loc[df['Pclass'] == 2 , 'Pclass'] = 'Gold'
df.loc[df['Pclass'] == 3 , 'Pclass'] = 'Bronze'
```

```python
df.loc[df['Name'].str.startwith('') , 'Pclass'] = 'Bronze'
```

```python
df.head()
```

```python
df['Pclass'].value_counts()
```
